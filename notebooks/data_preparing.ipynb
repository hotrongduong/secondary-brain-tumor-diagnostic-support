```python
import os
import pandas as pd
import numpy as np
```

```python
manifest_file_path = "../data/brisc2025/manifest.csv"
output_dir = "../data/brisc2025"

manifest_df = pd.read_csv(manifest_file_path)
manifest_df.head()
```

```python
images_df = manifest_df[manifest_df['task'] == 'classification'].copy()
masks_df = manifest_df[manifest_df['is_mask'] == True].copy()
print(f"Tìm thấy {len(images_df)} ảnh và {len(masks_df)} mặt nạ.")

print("Đang taọ khóa tra cứu (lookup keys).....")
images_df['lookup_key'] = images_df['filename'].str.replace(".jpg", "", regex=False)
masks_df['lookup_key'] = masks_df['filename'].str.replace(".png", "", regex=False)

mask_path_dict = masks_df.set_index('lookup_key')['relative_path'].to_dict()
print("Đang ánh xạ mặt nạ vào các ảnh tương ứng.....")
images_df['mask_relative_path'] = images_df['lookup_key'].map(mask_path_dict)

print("Đang tạo DataFrame MTL cuối cùng.....")
df_mtl_final = images_df[[
    'relative_path',
    'mask_relative_path',
    'tumor_label',
    'split'
]].copy()
```

```python
train_df = df_mtl_final[df_mtl_final['split'] == 'train']
test_df = df_mtl_final[df_mtl_final['split'] == 'test']

train_output_path = os.path.join(output_dir, "train_clean.csv")
test_output_path = os.path.join(output_dir, "test_clean.csv")

train_df.to_csv(train_output_path, index=False)
test_df.to_csv(test_output_path, index=False)

train_df.head()
```
